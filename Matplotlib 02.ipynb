```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

__Membuat Plotting Sederhana__

```python
plt.plot([2, 5, 7, 11])
plt.ylabel('sumbu y')
plt.show()
```

```python
plt.plot([2, 3, 4, 7], [5, 3, 2, 16])
plt.show()
```

__Pengaturan Format Pada Plot__

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'g*')
plt.axis([0, 6, 0, 20])
plt.show()
```

Selain menyertakan 2 buah deret bilangan sebagai parameter untuk sumbu x dan y, terdapat parameter ke-3 yang bisa disertakan untuk mengatur warna dan jenis plotting

__Multi Plot Dalam Satu Pemanggilan Fungsi Plot()__

```python
t = np.arange(0., 5.,0.2)
t
```

```python
plt.plot(t, t, 'r--',
        t, t**2, 'bs',
        t, t**3, 'g^')
plt.show()
```

__Plotting Dengan Keywords__

```python
data = {'a': np.arange(50),
       'c': np.random.randint(0, 50, 50),
       'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

data
```

Menampilkan data dictionary

```python
plt.scatter('a', 'b', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

__Ploting Untuk Tipe Data Categorial__

Matplotlib juga mendukung plotting untuk data dengan tipe katergori atau biasa dikenal dengan istilah cageorical data atau categorical variable

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]
```

```python
plt.plot(names, values)
plt.show()
```

```python
plt.scatter(names, values)
plt.show()
```

```python
plt.bar(names, values)
plt.show()
```
